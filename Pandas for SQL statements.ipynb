### SELECT DISTINCT engine FROM planes

### SELECT DISTINCT type, manufacturer, model, seats, engine FROM planes

```python
sql_result = pd.read_sql_query("SELECT DISTINCT type, manufacturer, model, seats, engine FROM planes", con)
sql_result
```

```python
planes[["type", "manufacturer", "model", "seats", "engine"]].drop_duplicates().reset_index(drop=True)
```

### SELECT COUNT(*), engine FROM planes GROUP BY engine

```python
sql_result = pd.read_sql_query("SELECT COUNT(*), engine FROM planes GROUP BY engine", con)
sql_result
```

```python
planes.groupby("engine").count()
```

```python
df = pd.DataFrame(planes.groupby("engine").count()["manufacturer"]).reset_index()[["manufacturer","engine"]]
df.columns = ["COUNT(*)","engine"]
df
```

alternatively:

```python
pd.DataFrame(planes.engine.value_counts())
```

### SELECT COUNT(*), engine, type FROM planes GROUP BY engine, type

```python
sql_result = pd.read_sql_query("SELECT COUNT(*), engine, type FROM planes GROUP BY engine, type", con)
sql_result
```

```python
df = pd.crosstab(planes.engine, planes.type).unstack().reset_index()
df.columns = ["type", "engine", "COUNT(*)"]
df[df["COUNT(*)"]>0]
```

### SELECT MIN(year), AVG(year), MAX(year), engine, manufacturer FROM planes GROUP BY engine, manufacturer

```python
sql_result = pd.read_sql_query("SELECT MIN(year), AVG(year), MAX(year), engine,\
                                manufacturer FROM planes GROUP BY engine, manufacturer", con)
sql_result.head(8)
```

```python
df = planes.groupby(["engine", "manufacturer"]).agg(['min', 'mean', 'max'])["year"].reset_index()
df = df[["min","mean","max","engine","manufacturer"]]
df.head(8)
```

### SELECT manufacturer, COUNT(*) AS howmany FROM planes GROUP BY manufacturer ORDER BY howmany

```python
sql_result = pd.read_sql_query("SELECT manufacturer, COUNT(*) AS howmany FROM planes GROUP BY manufacturer ORDER BY howmany", con)
sql_result
```

```python
df = pd.DataFrame(planes["manufacturer"].value_counts(ascending=True)).reset_index().sort_values(by=["manufacturer", "index"])
df.columns = ["manufacturer", "howmany"]
df
```
